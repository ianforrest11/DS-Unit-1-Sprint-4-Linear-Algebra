## 1.1 Create a two-dimensional vector and plot it on a graph

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
# create vector/array with coordinates (10,10,10), starting from point (0,0,0)
tdv = np.array([0,0,0,10,10,10])

# create zip
X, Y, Z, U, V, W = zip(tdv)

# plot red 3d vector
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, color = 'red')
ax.set_xlim([0, 15])
ax.set_ylim([0, 15])
ax.set_zlim([0, 15])
plt.show();
```

```
# define matrices
A = [[1,2],
     [3,4],
     [5,6]]

B = [[2,4,6]]

C = [[9,6,3],
     [4,7,11]]

D = [[1,0,0],
     [0,1,0],
     [0,0,1]]

E = [[1,3],
     [5,7]]

# A can be multiplied with C and E
# B can be multiplied with A and D
# C can be multiplied with A and D
# D can be multiplied with A
# E can be multiplied with C
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
# define new matrices
CD = np.matmul(C,D)
AE = np.matmul(A,E)
BA = np.matmul(B,A)
print(CD) # resulting matrix is 2x3, which are the same dimensions as matrix C
print(AE) # resulting matrix is 3x2, which are the same dimensions as matrix A
print(BA) # resulting matrix is 1x2, which are the same row dimensions as matrix B and same column dimensions as matrix A
```

## 3.1 Find $IG$ (be sure to show your work) 😃

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\end{align}

```
# define G and Ginv
G = ([[12,11],
      [7, 10]])

Ginvcoeff = 1/((12*10) - (11*7))
Ginvarr = ([[10,-7],
            [-11,12]])
Ginv = np.dot(Ginvcoeff, Ginvarr)

# multiply G by its inverse
np.dot(G,Ginv)
```
